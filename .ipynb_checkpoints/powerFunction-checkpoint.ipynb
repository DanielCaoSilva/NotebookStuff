```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits import mplot3d
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
matplotlib.rcParams['figure.figsize'] = (10, 5)
matplotlib.rcParams['figure.dpi'] = 200
```

```python
n = np.linspace(10,100,90)
theta = np.linspace(10,12,90)
#theta = 10.1
p100 = 1-(10/theta)**(100)*(0.95)
p10 = 1-(10/theta)**(10)*(0.95)
```

```python
plt.plot(theta,p10)
```

```python
plt.plot(theta,p100)
```

```python
X,Y = np.meshgrid(n,theta)
Z = np.zeros(shape=(90,90))
for i in range(90):
    Z[:,i]=(1-(10/theta)**n[i] * 0.95)
```

```python
fig = plt.figure() 
axes = fig.gca(projection ='3d') 
axes.plot_surface(X,Y,Z)   
plt.title('Power Function: n=10:100,theta=10:12')
plt.xlabel('Sample Size')
#plt.ylabel('Theta')
#plt.zlabel('Power Function')
axes.set_xlabel('Sample Size')
axes.set_ylabel('Theta')
axes.set_zlabel('Power Function')
plt.show()
```
