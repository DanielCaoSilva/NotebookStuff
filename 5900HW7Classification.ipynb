## Problem 2

```python
import pandas as pd 
import numpy as np 
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

y: student passed final exam (1 if yes, 0 if no) 

x1: hours studied for final exam 

x2: hours slept night prior 

```python
x1 = np.array([2,6,6,5,2])
x2 = np.array([5,5,9,6,8])
e = np.exp(-6.25+0.7*x1 + 0.5*x2)
p = e/(1+e)
yhat = (p>0.5).astype(int)
y = np.array([0,1,1,1,0])
dfTest = pd.DataFrame({'y':y,'x1':x1,'x2':x2,'p':p,'yhat':yhat})
print(dfTest.to_markdown(tablefmt="grid"))
cfMat = confusion_matrix(y,yhat)
cfDf = pd.DataFrame(data=cfMat, index = ["Negative","Positive"], columns = ["Negative","Positive"])
print(cfDf)
acc = (cfMat[1,1] + cfMat[0,0])/(cfMat[1,1]+cfMat[0,0]+cfMat[0,1]+cfMat[1,0])
print("Acc:",acc)
```
